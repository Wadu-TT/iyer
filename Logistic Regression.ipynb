```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def gradient_descent(df , w , b , a , iter):
    x = np.array(df.iloc[: , 0:9])
    y = np.array(df.iloc[: , 9])
    m= x.shape[0]
    for i in range(iter):
            # Compute predictions
            y_hat = sigmoid(np.dot(x, w) + b)

            # Compute gradients
            dw = np.dot(x.T, (y_hat - y)) / m
            db = np.sum(y_hat - y) / m

            # Update weights and bias
            w -= a * dw
            b -= a * db
        
    return w, b
```
